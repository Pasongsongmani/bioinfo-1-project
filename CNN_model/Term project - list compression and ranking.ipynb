```python
import pandas as pd
control_binding = pd.read_csv("control_binding_site_upper.csv", sep=',')
#non_binding = pd.read_csv("non_binding_site_upper.csv", sep=',')
binding = pd.read_csv("putative_binding_site_upper.csv", sep=',')
```

```python
control_binding
```

```python
control_binding_seq = []

for i in range(142385):
    control_binding_seq.append(control_binding.loc[i][2])
    
control_overlaped_seq = []
temp = []

for seq in control_binding_seq:
    if seq in temp:
        pass
    else:
        temp.append(seq)
        control_overlaped_seq.append([seq, control_binding_seq.count(seq)])

import pandas as pd
control_overlaped_seq = pd.DataFrame(control_overlaped_seq)

control_overlaped_seq['Counting_Rank'] = control_overlaped_seq[1].rank(ascending = False)
control_overlaped_seq = control_overlaped_seq.set_index('Counting_Rank')
control_overlaped_seq = control_overlaped_seq.sort_index()

control_overlaped_seq
```

```python
control_overlaped_seq.to_csv("control_entropy_binding_motif")
```
